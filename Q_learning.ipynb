```python
from maze import *

env = Maze()
env.snum   # 112
env.anum   # 4
env.slip   # 0.1
initial_state = env.reset()
state = initial_state
```

```python
import numpy as np

q_lookup = np.zeros((env.snum, env.anum))
```

```python
alpha = 0.1          # learning rate
gamma = 0.9          # discount factor
epsilon = 0.1        # exploration rate
num_episodes = 5000

for episode in range(num_episodes):
    state = np.random.randint(env.snum)  # randomize start to cover all flag-state combos
    done = False

    while not done:
        # ε-greedy action selection
        if np.random.rand() < epsilon:
            action = np.random.randint(env.anum)
        else:
            action = np.argmax(q_lookup[state])

        reward, next_state, done = env.step(state, action)

        # Q-learning update
        q_lookup[state, action] += alpha * (
            reward + gamma * np.max(q_lookup[next_state]) - q_lookup[state, action]
        )

        state = next_state
```
